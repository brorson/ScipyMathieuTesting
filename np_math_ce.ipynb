```python
def tridiagonal_matrix(n, a, b, c):
    M = np.zeros((n, n))
    print(f'shape a = {a.shape}, shape b = {b.shape}, c shape = {c.shape}')
    for i in range(n):
        M[i][i] = b[i]
        if i > 0:
            M[i][i-1] = a[i]
        if i < n - 1:
            M[i][i+1] = c[i]

    return M
```

```python
def make_matrix_e(q, vn):
    N = len(vn)
    h = vn[1]-vn[0]
    c = 2*q*np.cos(2* vn).reshape((len(vn),1))
    hm2 = np.ones((N,1))/(h*h)
    l = -2*hm2-c
    print(f'h = {h}')
    print(f'hm2 = {hm2}')
    print(f'-2*hm2-c = {l}')
    print(f'c ={c}')
    A = tridiagonal_matrix(N, hm2, -2*hm2-c, hm2)
    A[0,1] = 2*hm2[0]
    A[N-1,N-2] = 2*hm2[0]
    print(A)
    return A
```
